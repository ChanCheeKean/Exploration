```python
import wikipediaapi
import pandas as pd
from urllib.parse import unquote
import time
```

```python
wiki_wiki = wikipediaapi.Wikipedia(
    language='en',
    extract_format=wikipediaapi.ExtractFormat.WIKI
)
```

```python
df_test = pd.read_csv('data/gwikimatch_test_set.tsv', sep='\t', names=['label', 'url_one', 'url_two'])
df_train = pd.read_csv('data/gwikimatch_train_set.tsv', sep='\t', names=['label', 'url_one', 'url_two'])
df_eval = pd.read_csv('data/gwikimatch_eval_set.tsv', sep='\t', names=['label', 'url_one', 'url_two'])
```

```python
lis_test_one = []
start = time.time()
for ind, title in enumerate(df_test['url_one'].apply(lambda x: x.split('/')[-1]).values):
    title = unquote(title)
    page_py = wiki_wiki.page(title)

    try:
      if page_py.exists():
          text = page_py.text
      else:
          text = None
          print(ind, title)
    except:
      text = None

    lis_test_one.append(text)
    if ind % 100 == 0:
        print(ind, title, round(time.time() - start, 2))
df_test['doc_one'] = lis_test_one
        
lis_test_two = []
start = time.time()
for ind, title in enumerate(df_test['url_two'].apply(lambda x: x.split('/')[-1]).values):
    title = unquote(title)
    page_py = wiki_wiki.page(title)
    try:
      if page_py.exists():
          text = page_py.text
      else:
          text = None
          print(ind, title)
    except:
      text = None
    lis_test_two.append(text)
    if ind % 100 == 0:
        print(ind, title, round(time.time() - start, 2))

df_test['doc_two'] = lis_test_two
df_test.to_csv('./data/wikidocpair_test.csv', index=False)
```
